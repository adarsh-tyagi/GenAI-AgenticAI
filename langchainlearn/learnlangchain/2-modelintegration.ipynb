### Streaming and Batch

#### Stream
Instead of printing response all at once, just print it as it is generating

```python
model.invoke("Write me 500 words essay on Databases ?")
```

```python
model.stream("Write me 500 words essay on Databases ?")
```

```python
for chunk in model.stream("Write me 500 words essay on Databases ?"):
    print(chunk.text, end="|", flush=True)
```

#### Batch
Collection of independent requests to a model, processing will be done parallely

```python
responses = model.batch([
    "why do parrots have colorful feathers ?",
    "What is gravity ?",
    "Who was ramanujan and what theorems he gave ?"
])
for response in responses:
    print(response.content)
```

```python
model.batch([
    "why do parrots have colorful feathers ?",
    "What is gravity ?",
    "Who was ramanujan and what theorems he gave ?"],
            config={
                'max_concurrency': 2, # limit to 2 parallel calls
            })
```
